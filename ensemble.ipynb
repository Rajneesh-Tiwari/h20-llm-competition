### Mistral and Zephyr ensemble

```python
mistral_sub = pd.read_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb005/submit.csv')
zephyr_sub = pd.read_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb006/submit.csv')
```

```python
ens = (mistral_sub.iloc[:,1:]+zephyr_sub.iloc[:,1:])/2
mistral_sub.iloc[:,1:] = ens
```

```python
idxmax = mistral_sub.iloc[:,1:].idxmax(1).values.tolist()
```

```python
for ix, max_val in enumerate(mistral_sub.iloc[:,1:].max(1).values.tolist()):
    if max_val>0.99:
        _idxmax = idxmax[ix]
        remaining_ids = [i for i in mistral_sub.iloc[:,1:].columns if _idxmax not in i]
        mistral_sub.loc[ix,remaining_ids] = 0.
        mistral_sub.loc[ix,_idxmax] = 1.
```

```python
mistral_sub.to_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb006/mistral_zephyr_ense_pp.csv',index=False)
```

### Weights

```python
mistral_sub = pd.read_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb005/submit.csv')
zephyr_sub = pd.read_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb006/submit.csv')
public_sub = pd.read_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/public_sub.csv')
```

```python
ens = (mistral_sub.iloc[:,1:]*0.35+zephyr_sub.iloc[:,1:]*0.55+public_sub.iloc[:,1:]*0.1)
mistral_sub.iloc[:,1:] = ens
```

```python
idxmax = mistral_sub.iloc[:,1:].idxmax(1).values.tolist()
```

```python
for ix, max_val in enumerate(mistral_sub.iloc[:,1:].max(1).values.tolist()):
    if max_val>0.99:
        _idxmax = idxmax[ix]
        remaining_ids = [i for i in mistral_sub.iloc[:,1:].columns if _idxmax not in i]
        mistral_sub.loc[ix,remaining_ids] = 0.
        mistral_sub.loc[ix,_idxmax] = 1.
```

```python
mistral_sub.to_csv('///mnt/c/Personal/Competitions/Kaggle/h2oai-predict-the-llm/runs/nb006/mistral_zephyr_publicsub_weighted_ense_pp.csv',index=False)
```
